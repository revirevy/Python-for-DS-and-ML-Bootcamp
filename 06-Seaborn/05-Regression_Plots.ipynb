# 05 - Regression Plots 

```python
## Seaborn has many built-in capabilities for regression plots 

## We will only cover the lmplot() function -> Allows you to display linear models with linear models 
                                        ##  -> We will also learn how to size and change the aspect ratio 
                                        ##  for seaborn plots. 
import seaborn as sns 
%matplotlib inline 
tips = sns.load_dataset('tips')
tips.head()
```

```python
## Using lmplot 

sns.lmplot(x ='total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['v', 'o'], 
          scatter_kws = {'s':100}) ##Size of the scatterplot

## Returns a scatterplot with a linear fit on it 
## We can also pass a hue parameter -> We can infer some conclusions from our data -> Both males and females tip 
                                ##     similarly based on the total_bill parameter 
    
## We can pass matplotlib style parameters, e.g. markertypes
## To affect matplotlib directly (for this case) we can use the scatter_kws parameter -> Note we can always 
## check the seaborn documentation and also, you won't be customizing this much with seaborn anyways. 
```

```python
## We may also use a Grid instead of separating by hue. Just pass in a col parameter and/or row parameter.
## hue can also be added, of course.

sns.lmplot(x ='total_bill', y = 'tip', data = tips, col = 'day', hue = 'sex')
```

```python
##Due to the size of the above plot, we must modify its aspect and size ratio: 

##We can do so with the aspect and height parameter: 

sns.lmplot(x ='total_bill', y = 'tip', data = tips, col = 'day', hue = 'sex', aspect = 0.6, height = 8 )

## In the next lecture, we will control the size and style of other elements, such as fonts!
```
