# 06 - Style and Color

```python
## We are going to discuss how to control style and color in seaborn in a more formal manner: 

import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
### A - Setting particular styles 

sns.set_style('ticks') ##Control the style 
sns.countplot(x = 'sex', data = tips)
sns.despine(left = True, bottom = True) ##Left and bottom and by default False to be removed, whereas Right and Top
                                        ##are True default to be removed.
```

```python
## B -Size and aspect

## We can control size and aspect using matplotlib plt.figsize() to control most sizes of seaborn plots, e.g.

plt.figure(figsize = (12,3)) ##Using matplotlib to control figure and aspect of our plots 
sns.countplot(x = 'sex', data = tips)
```

```python
## Control of seaborn grids can also be obtained by passing the size and aspect parameters, e.g. 

## Scale and context -> Seaborn comes with the convenient set_context() which allows you to overwrite default parameters
## and change font scales depending on the purpose of your figure, e.g. a poster: 

sns.set_context('notebook', font_scale = 2) ## font_scale = 3 - 3 times the deault font size
sns.countplot(x = 'sex', data = tips)
```

```python
## Palettes and colors 

sns.set_context('notebook')
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue= 'sex', palette = 'BuGn')

## Go to the following link for more colormaps! : 
```
