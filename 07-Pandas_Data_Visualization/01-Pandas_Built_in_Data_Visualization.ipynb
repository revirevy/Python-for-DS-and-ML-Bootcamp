# 01-Pandas Built-in Data Visualization

```python
### In this lecture we are going to explore Pandas built-in Data Visualization capabilities which are 
### built directly from matplotlib. Finally, we can call data visualization methods directly from Data Frames 

import numpy as np 
import pandas as pd 
%matplotlib inline 

df1 = pd.read_csv('df1', index_col = 0)
```

```python
df1.head()
```

```python
df1.info()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

```python
### Examples 


##Getting a histogram of all values in column A in df1 

df1['A'].hist(bins = 30) ## Note that this method call calls matplotlib under the hood 
                         ## We are also able to call matplotlib and seaborn commands to add to our plot 
```

```python
import seaborn as sns ##Previous plots will be displayed with seaborn style plots 
```

```python
df1['A'].hist(bins = 30)
```

```python
### Pandas has several plot types built into it. 

## We will see several examples, pretty much all of the plots can be called from the Data Frames. 

##Calling from a column 

df1['A'].plot(kind = 'hist', bins = 30)
```

```python
### We can call hist directly off of a plot -> Preferred method 

df1['A'].plot.hist(bins = 50)
```

```python
### Let's check out some examples: 

## Ex. 1
df2.plot.area(alpha = 0.5)
```

```python
## Ex. 2
df2.plot.bar() ##Takes index value (must be categorical) and will list the values per each column 
```

```python
## We can call the stacked parameter as follows: 

p = df2.plot.bar(stacked = True)
p.legend(loc = 0)
```

```python
## Ex. 3 
## Line plots - Everything you learned from matplotlib can be applicable to these plots 
df1.plot.line( y = 'B', figsize = (10,4), lw = 0.5)
```

```python
## Ex. 4 
## Scatter Plots 

df1.plot.scatter(x= 'A', y = 'B', c = 'C', cmap = 'BuGn') ##The c parameter is the color - Shows an additional level of information
                                                          ##The s parameter is the Data Frame column 
```

```python
df1.head()
```

```python
df1.plot.scatter(x = 'A', y = 'B', s = df1['C']*10)  ## We can also pass a size argument to show this additional level of information 
```

```python
### Always check out the documentation in case of doubt! 

## Ex.5

##Boxplots 

df2.plot.box(figsize = (10,10))
```

```python
### Ex. 6

## For bivariate data we can use a hexagonal binary plot 

df = pd.DataFrame(np.random.randn(1000,2), columns = ['a', 'b'])
df.head()
```

```python
df.plot.hexbin(x = 'a', y = 'b', gridsize = 25, cmap = 'coolwarm')

## gridsize -> Controls the size of the hexagons                                                                
## Hexagons get darker as they have more points within them. 
```

```python
## Ex 7. 

### KDE Plots 

df2['a'].plot.kde() ## We can also call density() instead of kde()
```

```python
### You can also call the kde() function on the whole Data Frame 

df2.plot.density()
```
