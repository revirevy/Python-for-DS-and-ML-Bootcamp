```python
##INTRODUCTION
### Brief introduction to matplotlib 

# -Matplotlib is the most popular plotting library for Python 
# -It gives you control over every aspect of a figure 
# -It was designed to have a similar feel to MatLab's graphical plotting

##Works very well with numpy and pandas libraries. Later -> Seaborn which is based on Matplotlib
##Use the following link to see a gallery of matplotlib examples: 

##https://matplotlib.org/gallery/index.html
```

```python
##Matplotlib Part I 

##Importing matplotlib library 
import matplotlib.pyplot as plt
```

```python
##When using the Jupyter notebook: 

##Allows you to see the plots you create using the Jupyter notebook 
%matplotlib inline 

##If using other environments use the plt.show() method after every plot code 

import numpy as np 

x = np.linspace(0,5,11)
y = x**2
```

```python
##Basic matplotlib allows you to create a plot very quickly. We can do so in two ways: 

## A- Functional method 
plt.plot(x,y, 'r-') ##We can also add color and line styles like in MATLAB

##Adding an x, y label and a title to our plot
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')

plt.show() ##Prints the plot instead  of showing it -> Not necessary for JP notebook
```

```python
###Creating multiplots in the same canvas 

##Use plt.subplot()
plt.subplot(1,2,1) ##number of rows, number of columns, plot we are referring to 

plt.plot(x,y, 'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
### B - Object oriented matplotlib 

##More formal -> Uses figure objects and calls methods upon them 

fig = plt.figure() #Instantiating a plot object - Think of this object as a blank canvas 

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) ##Takes four arguments: Left, Bottom, Length and Width arguments as percentages 

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure() ## Main idea of matplotlib -> Create an empty canvas (figure object), later, 
                    ## add axes and other details.

##An insert set of axes 
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('LARGE PLOT')
axes2.set_title('SMALL PLOT')
```
