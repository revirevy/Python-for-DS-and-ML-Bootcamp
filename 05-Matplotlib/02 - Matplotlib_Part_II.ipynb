```python
import matplotlib.pyplot as plt
%matplotlib inline 

import numpy as np 

x = np.linspace(0,5,11)
y = x**2
```

```python
##Creating subplots using the Object oriented method 
fig, axes = plt.subplots(nrows = 1, ncols= 2) ## You may specify the number of rows and columns (Number of subplots
                                              ## subplots() -> An axis manager on top of plt.figure()   

plt.tight_layout() ##Fixes the overlap issue (Try nrows = 3, ncols = 3)

##Iterating through our axis

# for curr_axis in axes:
#    curr_axis.plot(x,y)

### Since we can iterate through our axes object, we can also index it:
axes[0].plot(y,x)
axes[0].set_title('First Plot')

axes[1].plot(x,y)
axes[1].set_title('Second Plot')
```

```python
axes ##An np.array of matplotlib axes 
```

```python
## Figure size and DPI 

## Matplotlib allows you to specify each of these aspects when you care instantiating the figure object 

fig = plt.figure(figsize = (8,2), dpi = 100)  ##dpi = pixels per inch -> Default works ok

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
## We can do something similar with subplots 

fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize =(8,4))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
## Saving a figure 
## We can use matplotlib to output high quality plots in a gat number of formats including ong, jpeg, pdf, etc...

fig.savefig('my_picture.png', dpi =200) ###Specify name.extension
```

```python
##Using and adding legends
fig2 = plt.figure()

ax = fig2.add_axes([0,0,1,1])

ax.plot(x,x**2, label = 'X squared') 
ax.plot(x,x**3, label = 'X cubed')

ax.set_title('Title')
ax.set_xlabel('X label')
ax.set_ylabel('Y label')

ax.legend(loc=0) ## Need to add string labels to each plot call such that the legend() function may reference
                 ## We cay specify the location of the legend in case that any overlpa happens -> 0 = best
                 ## Check out https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.legend.html
        
                 ##If each of the above location options does not satisfy our needs we can determine a location 
                 ## using a 2-tuple and declaring where our legend should be
```
