```python
### In this lecture -> Setting colors, line widths, line types and customizing our plot appeareance. 

import matplotlib.pyplot as plt
%matplotlib inline 

import numpy as np 

x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'purple', lw = 3, alpha =0.5, ls = '--', marker = 'o', markersize = 20,
       markerfacecolor = 'blue', markeredgewidth = 3, markeredgecolor = 'green')
                                             ##Color argument can take multiple parameters 
                                             ## Color name as strings 
                                             ## RGB hexcodes as strings -> Consider using an online tool for custom colors 
        
                                             ## Linewidth argument -> We must pass in an int - shortcut -> Type lw
                                             ## Alpha argument ->  Allows you to control how transparent your line is
                
                                             ## Line Style -> Quite a few options that define the style of the line 
                                             ## ':'. '--' , '-' are possible options - shortcut -> Type ls
                        
                                             ## Markers -> Are used for when we have a few data points. 
                                             ## We pass some marker symbols as strings such as 'o' that mark out each
                                             ## point on x vs y in your plot. 
                                    
                                             ## Markersize -> Pass in an int as a parameter to specify marker size
                                        
                                             ## Marker appeareance -> Specify markerfacecolor, markeredgewidth, markeredgecolor
                            
```

```python
## Controlling plot range 

fig2 = plt.figure()

ax = fig2.add_axes([0,0,1,1])
ax.plot(x,y,color='purple', lw = 2, ls = '--')

##To control the range in the x-axis: 
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
